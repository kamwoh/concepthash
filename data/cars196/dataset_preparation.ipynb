```python
import scipy.io as sio

train_mat = sio.loadmat('stanford_cars/devkit/cars_train_annos.mat', squeeze_me=True)
train_fn2class = {annotation['fname']: annotation['class'] - 1 for annotation in train_mat['annotations']}

test_mat = sio.loadmat('stanford_cars/cars_test_annos_withlabels.mat', squeeze_me=True)
test_fn2class = {annotation['fname']: annotation['class'] - 1 for annotation in test_mat['annotations']}
```

```python
class_names = sio.loadmat('stanford_cars/devkit/cars_meta.mat', squeeze_me=True)["class_names"].tolist()
c2i = {cls: i for i, cls in enumerate(class_names)}
```

```python
class_names
```

```python
with open('train.txt', 'w+') as f:
    for fn in train_fn2class:
        path = f'data/cars196/stanford_cars/cars_train/{fn} {train_fn2class[fn]}'
        f.write(path + '\n')

with open('test.txt', 'w+') as f:
    for fn in test_fn2class:
        path = f'data/cars196/stanford_cars/cars_test/{fn} {test_fn2class[fn]}'
        f.write(path + '\n')

with open('class_names', 'w+') as f:
    for c in class_names:
        f.write(c + '\n')
```
